### boolen 
#### are objects which store binary values / sometimes reffered to as True/False

```python
my_str = "Diwakar"
```

```python
print(my_str.isalnum()) # Returns True if all charecters are numbers and 
print(my_str.isalpha()) # Returns True if it has all char values
print(my_str.islower()) # check for lowercase strings
print(my_str.istitle()) # checks if First char is ALpha
print(my_str.isupper()) # test upper case strings
print(my_str.startswith("D")) # True if it starts
print(my_str.endswith("r"),"7") # True if it ends
```

# Lists 
- mutable and indexed data structure 
- can store multiple data types of data

```python
lst_example = []
print(type(lst_example))
type(lst_example)
```

```python
lst = list() # 2nd way of creating list
type(lst)
```

```python
lst_2 = ["diwakar babu",80,"balu",79]
print(type(lst_2))
print(lst_2)
```

## Addition/manpulating of items in list

```python
lst_2.append("babu") # adds at last 
print(lst_2)
```

### inseritng at specified index

```python
lst_2.insert(2,"Yoga") # adds at specified index
print(lst_2)
```

### Inserting multiple values 

```python
lst_2.append(["manish","amit"]) # adds by creating nested lists 
print(lst_2)
lst_2.extend(["Raju","shyam"]) # adds multiple values at last
print(lst_2)
```

```python
lst = [1,2,3,5]
print(lst)
lst.pop()
print(lst)
lst.pop(1)
print(lst)
```

```python
print(sum(lst))
print(max(lst))
print(min(lst))
print()
lst.extend([4,5,6,7])
print(lst)
      
```

```python
print(lst.count(1)) # prints occurance of value in list
```

```python
lst.index(3,1,7) # 3 is value and (1 and 9 are start value and end value range which it will check)
```

### Deleting elements in a list

```python
print(lst)
lst.remove(4)
print(lst)
```

```python
print(lst)
lst.pop(4)
print(lst)
```

# Sets
- sets unordered and unindexed mutable 
- iterable stores non-duplicate values

```python
set_example = {1,2,2,3,33,4,4,5}
print(set_example)
```

```python
lst = [val for val in set_example]
print(*lst)
print(type(lst))
```

# Dictionary
- ordered, indexed
- have keys and values

```python
dict = {1:"diwakar",2:"babu",3:"Shyam"}
```

```python
for i in dict.values():
    print(i)
```

```python
for i in dict:
    print(i)
```

```python
for i in dict.items():
    print(i)
```

```python
dict[4] = "Krishna"
for i in dict.values():
    print(i)
```

```python
print(dict[3])
```

## Nested Dictionaries

```python
car1_model = {"BmW":1960}
car2_model = {"fortuner":2000}
car3_model = {"Bolero":1876}

cars = {"car1":car1_model,"car2":car2_model,"car3":car3_model}
```

```python
print(cars)
print(cars["car1"])
print(cars["car1"]["BmW"])
```
